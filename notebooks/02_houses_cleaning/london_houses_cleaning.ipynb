```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/Dane/dane/london_cleaned.csv')
```

```python
neighbourhoods = data['neighbourhood_cleansed'].unique()
path = '/content/drive/MyDrive/Dane/hpm_la_2024/'
neighbourhoods
```

```python
for neighbourhood in neighbourhoods:
  neighbourhood = neighbourhood.replace(' ', '_')
  temp = pd.read_csv(path + neighbourhood + '_link_26122024.csv')
  temp = temp.drop(columns=['lad23cd','transactionid','lmk_key','CURRENT_ENERGY_EFFICIENCY','POTENTIAL_ENERGY_EFFICIENCY','CONSTRUCTION_AGE_BAND','classt'])
  temp['numberrooms'] = temp['numberrooms'].fillna(temp['numberrooms'].median())
  temp['neighbourhood'] = neighbourhood.replace('_', ' ')
  temp['propertytype'] = temp['propertytype'].str.replace('D', 'Wolnostojący', regex=False)
  temp['propertytype'] = temp['propertytype'].str.replace('S', 'Bliźniak', regex=False)
  temp['propertytype'] = temp['propertytype'].str.replace('T', 'Szeregowy', regex=False)
  temp['propertytype'] = temp['propertytype'].str.replace('F', 'Mieszkanie', regex=False)
  temp['duration'] = temp['duration'].str.replace('F', 'Własność', regex=False)
  temp['duration'] = temp['duration'].str.replace('L', 'Dzierżawa', regex=False)
  temp = temp[temp['priceper'].notna()]
  numeryczne = temp.select_dtypes(include='number')
  for col in numeryczne:
    temp[col] = temp[col].astype(str)
    temp[col] = temp[col].str.replace('.', ',', regex=False)
  temp = temp[temp['year'] >= '2020']
  temp.to_csv('london_houses_cleaned.csv', mode='a', header=False)
  print("Done")
```
